```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/arienugroho050396/Customer-Personality-Analysis/main/marketing_campaign.csv'
customer = pd.read_csv(url, sep='\t')
customer.head()
```

```python
print(customer.info())
```

```python
print(customer.isnull().sum())
```

```python
#checking duplicated data
customer.duplicated().sum()
```

```python
# 2. Handle Missing Values (specifically in 'Income')
print(f"Missing income values: {customer['Income'].isnull().sum()}")
```

```python
# Impute income with median value
median_income = customer['Income'].median()
customer['Income'] = customer['Income'].fillna(median_income)
```

```python
print(f"Missing income values: {customer['Income'].isnull().sum()}")
```

```python
#Convert Date Columns to Datetime
customer['Dt_Customer'] = pd.to_datetime(customer['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
```

```python
#Rename Columns to Snake Case
customer.columns = customer.columns.str.lower().str.replace(' ', '_')
```
